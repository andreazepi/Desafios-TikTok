Mostre todos os produtos entregues

```python
pedidos = [
    ('Camisa', 'Vestuário', 59.90, 'Entregue'),
    ('Tênis', 'Calçado', 199.90, 'Cancelado'),
    ('Calçado', 'Vestuário', 89.90, 'Entregue'),
    ('Fone de Ouvido', 'Eletrônicos', 129.90, 'Pendente'),
    ('Meia', 'Vestuário', 19.90, 'Entregue'),
    ('Notebook', 'Eletrônicos', 3299.90, 'Entregue'),
    ('Jaqueta', 'Vestuário', 139.90, 'Cancelado')
]

#mostre todos os produtos entregues
print('PRODUTOS ENTREGUES'.center(40, '-'))
for i in pedidos:
  if i[3] == 'Entregue':
      print(f'O produto {i[0]} está com o Status Entregue')
```

Vamos analisar o código passo a passo:
análise com chat




1.   pedidos = [...]: Esta linha inicializa uma lista chamada pedidos. Cada item dentro da lista é uma tupla, e cada tupla representa um pedido. As tuplas contêm informações como nome do produto, categoria, preço e status.












2.   print('PRODUTOS ENTREGUES'.center(40, '-')): Esta linha imprime um cabeçalho.

*   'PRODUTOS ENTREGUES' é a string que será exibida.
*   .center(40, '-') é um método de string que centraliza o texto 'PRODUTOS ENTREGUES' em uma largura total de 40 caracteres, preenchendo os espaços restantes com o caractere '-'. Isso serve para formatar a saída e torná-la mais legível, criando um título visualmente agradável.




3.   for i in pedidos:: Esta linha inicia um laço for. Ele itera sobre cada elemento da lista pedidos. Em cada iteração, a variável i assume o valor da tupla atual da lista (por exemplo, na primeira iteração, i será ('Camisa', 'Vestuário', 59.90, 'Entregue')).


4.   if i[3] == 'Entregue':: Dentro do laço, esta linha verifica uma condição.


*   i[3] acessa o quarto elemento da tupla i (lembre-se que a indexação começa em 0). Neste caso, i[3] corresponde ao status do pedido (por exemplo, 'Entregue', 'Cancelado', 'Pendente').
*   == 'Entregue' compara o status do pedido com a string 'Entregue'. A condição só é verdadeira se o pedido estiver com o status 'Entregue'.



5.   print(f'O produto {i[0]} está com o Status Entregue'): Esta linha só é executada se a condição if for verdadeira (ou seja, se o pedido estiver 'Entregue').


*   f'...' indica uma f-string (formatted string literal), que permite incorporar expressões Python diretamente dentro de strings prefixadas com f.

*   {i[0]} dentro da f-string acessa o primeiro elemento da tupla i, que é o nome do produto (por exemplo, 'Camisa', 'Meia', 'Notebook').


*   O resultado final é uma mensagem formatada que indica qual produto está com o status 'Entregue'.

Em resumo, o código percorre uma lista de pedidos, filtra aqueles que têm o status 'Entregue' e imprime uma mensagem personalizada para cada um desses produtos.

### Atividade Sugerida: Calcular o Valor Total dos Produtos Entregues

**Objetivo:** Calcule a soma dos preços de todos os produtos cujo status é 'Entregue' na lista `pedidos`.

```python
total_entregue = 0.0

pedidos = [
    ('Camisa', 'Vestuário', 59.90, 'Entregue'),
    ('Tênis', 'Calçado', 199.90, 'Cancelado'),
    ('Calçado', 'Vestuário', 89.90, 'Entregue'),
    ('Fone de Ouvido', 'Eletrônicos', 129.90, 'Pendente'),
    ('Meia', 'Vestuário', 19.90, 'Entregue'),
    ('Notebook', 'Eletrônicos', 3299.90, 'Entregue'),
    ('Jaqueta', 'Vestuário', 139.90, 'Cancelado')
]

print('PRODUTOS ENTREGUES'.center(40, '-'))
for i in pedidos:
  if i[3] == 'Entregue':
      print(f'O produto {i[0]} está com o Status Entregue')
      total_entregue += i[2]
      #Abreviação
      #jeito normal: total_entregue = total_entregue + i[2]
print(f'O valor total dos produtos entregues é: R$ {total_entregue:.2f}')
```
